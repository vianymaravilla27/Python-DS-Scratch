### Descenso de gradiente



El descenso de gradiente es un algoritmo de uso general que numéricamente encuentra un mínimo de funciones multivariables.

El descenso de gradiente es un algoritmo que estima numéricamente dónde una función genera sus valores más bajos. Eso significa que encuentra mínimos locales, pero no al establecer \nabla f = 0∇f=0del, f, equals, 0 como hemos visto antes. En lugar de encontrar mínimos manipulando símbolos, el descenso de gradiente aproxima la solución con números. Además, todo lo que necesita para ejecutarse es la salida numérica de una función, no requiere ninguna fórmula.

En particular, obtendremos de forma automática el mínimo de la función , y analizaremos el efecto que la tasa de aprendizaje tiene en el comportamiento del algoritmo.

En la figura de abajo se observa la función y, donde podemos verificar que el valor mínimo se obtiene cuando x = 0:

```python
import matplotlib.pyplot as plt
import numpy as np
import random as rd
#ejemplo del gradiente descendente aplicando a la funcion y**2 +1
#la ecuacion matematica para el gradiente (derivada) es 2*x

#se define aletoriamente

#x_inicial = rd.randint(1,100)
x_inicial = 5
print("La inicial es:" , x_inicial)
alpha = 0.4
#0.25 
n_iteraciones = 100

iteraciones = []
y = []

x = x_inicial
for i in (range(n_iteraciones)):
    print('.................')
    print('iteracion' , str(i+1))

    #Calcular la gradiente

    gradiente = 2*x
    
    #Actualizar "x" usando gradiente descendente
    x = x - alpha * gradiente
    #Almacenar iteracion y valor correspondinete
    y.append(x**2 + 1)
    iteraciones.append(i+1)

    #imprimir resultados
    print('x = ', str(x) , 'y =', str(x**2+1))
plt.subplot(1,2,1)
plt.plot(iteraciones,y)
plt.xlabel('iteracion')
plt.ylabel('y')

X = np.linspace(-5,5,100)
Y = X**2 + 1
plt.subplot(1,2,2)
plt.plot(X,Y,0.0,1.0,'ro')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

Aqui muestra el numero de iteraciones que se necesita para llegar a ese valor, necesita converger. 

si es plana el learning rate es demasiado grande para empezar a converger

si es 0.4 el learning rate es más rapido dado que nos acercamos más al valor de el punto 0
